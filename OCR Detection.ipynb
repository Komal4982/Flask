```python
class OCR:
    def __init__(self): 
        self.mylist=[]
        

    def readfile(self,path):  
       
        for file in glob.glob(path):
#             print(file)
            a=cv2.imread(file)
            img = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
            text=pytesseract.image_to_string(img)
#             print(text)
            # Initializing data variable
            name = None
            fname = None
            dob = None
            pan = None
            nameline = []
            dobline = []
            panline = []
            text0 = []
            text1 = []
            text2 = []


            # Searching for PAN
            lines = text.split('\n')
            for lin in lines:
                s = lin.strip()
                s = s.rstrip()
                s = s.lstrip()
                text1.append(s)

            #text1 = list(text1)
            text1 = list(filter(None, text1))
#             print(text1)

            # List Object Returned in the following order


            lineno=0 # to start from the first line of the text file.

            for wordline in text1:
                xx = wordline.split('\n')
                if ([w for w in xx if re.search('(INCOME|TAX|GOW|GOVT|GOVERNMENT|OVERNMENT|VERNMENT|DEPARTMENT|EPARTMENT|PARTMENT|ARTMENT|INDIA|NDIA)$', w)]):
#                     text1 = list(text1)
                    lineno = text1.index(wordline)
                    break

            text1 = list(text1)
            text0 = text1[lineno+1:]
#             print(text0)
            self.mylist.append(text0)
        return self.mylist 
    
    def extract_Name(self,readdata):
        Name=[]
        for i in readdata:
#             print(i)
            Name1=(i[0])
            Name.append(Name1)
        return Name
    
    
    def extract_Father_Name(self,readdata):
        FatherName=[]
        for k in readdata:
#             print(k)
            FatherName1=(k[1])
            FatherName.append(FatherName1)
        return FatherName

            
    def Birthdate_isValid(self,readdata):    
        Birthdate_list=[]
        for i in readdata:
            for j in i:
                Result=re.compile("([0-9]{2}\/[0-9]{2}\/[0-9]{4})") 
                Result1=Result.findall(j)
                Birthdate_list.append(Result1)
        return Birthdate_list

        
    def PanNo_isvalid(self,readdata): 
        PanNo_List=[]
        for j in readdata:
            for i in j:
                Result=re.compile("[A-Za-z]{5}\d{4}[A-Za-z]{1}") 
                Result1=Result.findall(i)
                PanNo_List.append(Result1)
#                 print(PanNo_List)
        return PanNo_List

        
def main():    
    obj=OCR()
    path="D:/Projects/repo/MLalgoandAnalysis/ImageProcessing/Image/*.*"
    readdata=obj.readfile(path)
#     print(readdata)
    
    Name=obj.extract_Name(readdata)
#     print(Name)
    
    
    FatherName=obj.extract_Father_Name(readdata)
#     print(FatherName)
    
#     df = pd.DataFrame(readdata, columns = ['Name', 'Father Name','BirthDate','aa','PanCard Number','bb','cc']).drop(['aa','bb','cc'],axis=1)
#     print(df)

    
    Birthdate=obj.Birthdate_isValid(readdata)
    Birthdate = [ele for ele in Birthdate if ele != []] 
#     print(Birthdate)
    
    PanNo=obj.PanNo_isvalid(readdata)
    PanNo = [ele for ele in PanNo if ele != []] 
#     print(PanNo)
    
#     

    OCR_PanCard = pd.DataFrame(
    {'Name':Name,'FatherName':FatherName,'Birthdate': Birthdate,
     'PanNo': PanNo
    })
    print(OCR_PanCard)    

    OCR_PanCard.to_csv("data.csv")
    
    
    
if __name__=='__main__':
    main()
```
